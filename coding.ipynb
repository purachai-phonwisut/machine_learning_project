```python
df1 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/Disposable_income_adj.csv', na_values=['NA'])
df2 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/values_long_format.csv', na_values=['NA'])

#changing United States to USA
df2['Country'].replace({'United States': 'USA'}, inplace=True)

#prepare df1 to merge
#renaming Geography to Country to match df2
df1.rename(columns={'Geography':'Country'}, inplace= True)

#defining numeric_column
df1_columns = ['2011', '2012', '2013', '2014', '2015', '2016', '2017']
df2_columns  = ['2011', '2013', '2015', '2017']

# converting string integer to float
df1[df1_columns] = df1[df1_columns].replace({',': ''}, regex=True).astype(float)
df2[df2_columns] = df2[df2_columns].replace({',': ''}, regex=True).astype(float)
# drop columns before merging


# df1_to_df2 data are droped to be merge with df2 and vice versa
df1_to_df2 = df1.drop(df1_columns + ['Category', 'Data Type', 'Unit', 'Currency Conversion', 'Current Constant'], axis=1)
df2_to_df1 = df2.drop(df2_columns, axis=1)
df1 = df1.drop(columns=['Category', 'Data Type', 'Unit', 'Currency Conversion', 'Current Constant'], axis=1)

# merge 
merge_df1 = pd.merge(df1, df2_to_df1, on='Country', how='inner')
merge_df2 = pd.merge(df2, df1_to_df2, on='Country', how='inner')

# Merge Final_df1 and Final_df2 on 'region' and 'Country'
# Add 'Total' column by summing columns 2 through 8
merge_df1['Total'] = merge_df1.iloc[:, 1:8].sum(axis=1).round(2)

# Calculate the average of the 'Total' column
merge_df1['avg disposible income per year (in million USD)'] = merge_df1['Total']/7
#Save a part of the table here
Final_df1 = merge_df1[['region', 'Country', 'avg disposible income per year (in million USD)']]

Final_df1

merge_df2['Total'] = merge_df2.iloc[:, 1:5].sum(axis=1)

# divide the sum by 4
merge_df2['avg unlicensed loss per year (in million USD)'] = merge_df2['Total']/4

# 
Final_df2 = merge_df2[['region', 'Country', 'avg unlicensed loss per year (in million USD)']]

Final_df2

# Merge Final_df1 and Final_df2 on 'region' and 'Country'
final_merged_df = pd.merge(Final_df1, Final_df2, on=['region', 'Country'], how='inner')
final_merged_df.sort_values(['region', 'Country'])
```
