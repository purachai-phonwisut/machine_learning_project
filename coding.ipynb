```python
import pandas as pd
df1 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/Disposable_income_adj.csv')
df2 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/values_long_format.csv')

df1.rename(columns={'Geography':'Country'}, inplace= True)
merged_df = pd.merge(df1, df2, on="Country")

merged2 = merged_df.iloc[:, [0,6,7,8,9,10,11,12]]
# merged2 = merged_df.loc[:, ['Country', '2017']]
merged2.rename(columns={'2011_x': '2011', '2013_x': '2013', '2015_x': '2015', '2017_x': '2017'}, inplace=True)

# merged2['avg'] = merged2['2011'] + merged2['2012']

numeric_columns = ['2011', '2012', '2013', '2014', '2015', '2016', '2017']
merged2[numeric_columns] = merged2[numeric_columns].replace({',': ''}, regex=True).astype(float)
merged2['avg'] = merged2.iloc[:, 1:8].sum(axis=1)
merged2

merged3 = merged2[['Country', 'avg']]

merged3
```
