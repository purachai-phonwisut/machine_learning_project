```python
#df1 = Disposable_income_adj.csv
#df2 = values_long_format.csv
df1 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/Disposable_income_adj.csv')
df2 = pd.read_csv('https://github.com/purachai-phonwisut/machine_learning_project/raw/main/Data%20cleaned/raw/values_long_format.csv')

#changing United States to USA
df2['Country'].replace({'United States': 'USA'}, inplace=True)

#prepare df1 to merge
#renaming Geography to Country to match df2
df1.rename(columns={'Geography':'Country'}, inplace= True)
#drop unnecessary columns(count=5)
df1.drop(columns=['Category','Data Type','Unit','Currency Conversion','Current Constant'], inplace=True)

#prepare df2 to merge
#drop years
df2.drop(columns=['2017', '2015', '2013', '2011'],inplace=True)

#selecting countries that exists in both df1 and 2 with merge
#joining 2 df with merge using innerjoin.
merge_df1 = pd.merge(df1,df2, on='Country', how='inner')

#defining numeric_column
numeric_columns = ['2011', '2012', '2013', '2014', '2015', '2016', '2017']
#converting string integer to float
merge_df1[numeric_columns] = merge_df1[numeric_columns].replace({',': ''}, regex=True).astype(float)

# Add 'Total' column by summing columns 2 through 8
merge_df1['Total'] = merge_df1.iloc[:, 1:8].sum(axis=1)

# Calculate the average of the 'Total' column
merge_df1['avg disposible income per year (in million USD)'] = merge_df1['Total']/7
#Save a part of the table here
Final_df1 = merge_df1[['region', 'Country', 'avg disposible income per year (in million USD)']]

Final_df1

# reminder to change nan to NA
```
