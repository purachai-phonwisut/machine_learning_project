```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data cleaned/AP_rate.csv")
df_melted = df.melt(id_vars='Country', var_name='Year', value_name = 'Value')
print(df_melted)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Year', y='Value', hue='Country')

plt.title('Country Data Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
